```python
def change_name(x):
    if "норм" in x or "сред" in x or "обыч" in x:
        x = x.replace(x, "норм")
    elif "плот" in x  or "пол" in x:
        x = x.replace(x, "круп")   
    elif "худощ" in x or "Xуд" in x or "xудощавое" in x:
        x = x.replace(x, "худ")   
    elif "Не указано" in x or "Ты объе" in x:
        x = x.replace(x, "-")  
    return x 
```
