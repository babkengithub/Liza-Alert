```python
list_links_users = list()
for i in range(len(list1)):
    list_links_users.append("https://lizaalert.org/forum/memberlist.php?start=")
#print(list_links)

for i in range(len(list1)):
    list_links_users[i] = "https://lizaalert.org/forum/memberlist.php?start=" + result[i]
```

```python
# Import all
import requests
import numpy as np
import pandas as pd
import re
import bs4
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
from scrapy import Selector
from nltk import word_tokenize
import nltk
nltk.download('punkt') #winlocker
from webbot import Browser 

nicks = []
chats = []
web = Browser()
web.go_to('https://lizaalert.org/forum/memberlist.php?start=0')
web.type('Researcher1' , into='Имя пользователя')
web.type('nikos007' , into='Пароль' , id='passwordFieldId') # specific selection
web.click('Вход' , tag='span') # you are logged in ^_^
for i in range(len(list_links_users)):
    web.go_to(list_links_users[i])
    a = web.get_page_source()
    
    while a.find('<td class="posts">0</td>') != -1:
        bec = a.find('<td class="posts">0</td>')
        a = a[0:(bec-125)] + a[(bec+50):]

    sel = Selector(text=a)
    course_as22 = sel.xpath('//tr[@class="bg1"]/td/a/text()').extract()
    course_as33 = sel.xpath('//tr[@class="bg2"]/td/a/text()').extract()
    
    for i in range(0, len(course_as22), 2):
        nicks.append(course_as22[i])

    for i in range(1, len(course_as22), 2):
        chats.append(course_as22[i])

    for i in range(0, len(course_as33), 2):
        nicks.append(course_as33[i])

    for i in range(1, len(course_as33), 2):
        chats.append(course_as33[i])
print(nicks)
print(chats)
```

```python
names_and_posts = dict(zip(nicks, chats))
```
