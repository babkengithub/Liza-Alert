```python
def transfer_to_weekday(x):
    xlist = list(map(int, x.split(".")))
    the_date = date(xlist[-1], xlist[-2], xlist[-3])
    if datetime.weekday(the_date) == 0:
        return "Понедельник"
    elif datetime.weekday(the_date) == 1:
        return "Вторник"
    elif datetime.weekday(the_date) == 2:
        return "Среда"
    elif datetime.weekday(the_date) == 3:
        return "Четверг"
    elif datetime.weekday(the_date) == 4:
        return "Пятница"
    elif datetime.weekday(the_date) == 5:
        return "Суббота"
    elif datetime.weekday(the_date) == 6:
        return "Воскресенье"
```

```python
transfer_to_weekday(a)
```

```python
df["День недели"] = df["Время пропажи"].apply(transfer_to_weekday)
```

```python
def transfer_to_season(y):
    if y in ["Январь", "Февраль", "Декабрь"]:
        return "Зима"
    elif y in ["Март", "Апрель", "Май"]:
        return "Весна"
    elif y in ['Июнь', 'Июль', 'Август']:
        return "Лето"
    else:
        return "Осень"
```

```python
df["Сезон"] = df["Месяц"].apply(transfer_to_season)
```
